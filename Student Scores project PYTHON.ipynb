```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"F:\Student Scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#Drop Unnamed column

df=df.drop('Unnamed: 0',axis=1)
print(df.head())
```

```python
# Change weekly study hours column

df['WklyStudyHours']=df['WklyStudyHours'].str.replace('05-Oct', ' 5-10')
df.head()
```

```python
# Gender distrubution
plt.figure(figsize=(5,5))
ax=sns.countplot(data =df,x='Gender')
ax.bar_label(ax.containers[0])
plt.title('Gender distrubution')
plt.show()
```

```python
#from the above chart we have analyst that:
# the number of females in the data is more than the number of males

gb=df.groupby("ParentEduc").agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb1)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent's Marital and Student's Score")
plt.show()
```

```python
#from the above we have concluded that there is no/negligible impact on the
#student's score due to their parent's marital status

sns.boxplot(data=df,x= 'MathScore' )
plt.show()
```
